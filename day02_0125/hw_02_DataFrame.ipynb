```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
file = '../subwaytime.csv'
file = pd.read_csv(file, skiprows = 2, header=None)
file
```

```python
station_df =  file[3]
people_df = file[35] + file[37]
```

```python
file['home'] = people_df
file['station'] = station_df
```

```python
file
```

```python
value = file['station'].unique()
value
```

```python
new_file = file.groupby('station')
new_file
```

```python
mydict = {}
for s in value:
    k = s
    v = new_file.get_group(s).sort_values(by = 'home', ascending = False)
    result = sum(v['home'])
    mydict[k] = result
```

```python
mydict
```

```python
list = sorted(mydict.items(), key = lambda x: x[1], reverse = True)
list[:5]
```

```python
newdict = {}
for i in range(5):
    key = list[i][0]
    value = list[i][1]
    newdict[key] = value
newdict
```

```python
x = newdict.keys()
y = newdict.values()
```

```python
plt.bar(x, y)
```
